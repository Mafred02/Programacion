Manejo de errores y excepciones

```python
usuarios = {
    "ANA": "1234",
    "JUAN": "abcd"
}

try:
    usuario = input("Usuario: ").upper()
    password = input("Contraseña: ")

    if usuario not in usuarios:
        raise KeyError("El usuario no existe")

    if usuarios[usuario] != password:
        raise ValueError("Contraseña incorrecta")

    else:
        print("Bienvenido,", usuario)

except KeyError as e:
    print("Error:", e)

except ValueError as e:
    print("Error:", e)
```

```python
try:
  num1 = (input("Ingrese el primer número: "))
  num2 = (input("Ingrese el segundo número: "))
  op = input("Ingrese la operación a realizar (+, -, *, /, %): ")

  if op == "+":
    resultado = num1 + num2
  elif op == "-":
    resultado = num1 - num2
  elif op == "*":
    resultado = num1 * num2
  elif op == "/":
    resultado = num1 / num2
  elif op == "%":
    resultado = num1 % num2
  else:
    raise ValueError("Operación no válida")

  print("El resultado de la operación es:", resultado)

except ValueError as e:
  print("Error de valor:", e)

except ZeroDivisionError as e:
  print("No se puede dividir entre cero")

except TypeError as e:
  print("Error de tipo:", e)
```

```python
productos = {
    "pan" : 1.5,
    "leche" : 2.0,
    "huevos" : 3.0
}

try:
  nombre = input("¿Que producto quieres?: ").lower()

  if nombre not in productos:
    raise KeyError("Producto no disponible")

  cantidad = int(input("¿Cuantos quieres?: "))

  if cantidad <= 0:
    raise ValueError("La cantidad debe ser mayor a 0")
  else:
    total = productos[nombre] * cantidad
    print("total a pagar:", total)

except ValueError as e:
  print("Error de valor: ", e)

except KeyError as e:
  print(e)
```

```python
try:

    datos = input("Ingresa notas separadas por coma: ")

    notas = [float(n) for n in datos.split(",")]

    if len(notas) == 0:
        raise ValueError("No ingresaste notas")

    promedio = sum(notas) / len(notas)

    posicion = int(input("¿Qué nota quieres ver (índice)?: "))

    if posicion < 0:
        raise ValueError("Ingresa solo índices positivos")

    print("Nota seleccionada:", notas[posicion])

    print("Promedio:", promedio)

except ValueError as e:
    print("Error de valor:", e)

except ZeroDivisionError:
    print("No se puede dividir entre cero")

except IndexError:
    print("Índice fuera de rango")
```

Expresiones regulares en Python

```python
import re
```

```python
texto = "Python es un lenguaje"
resultado = re.search("Python", texto)
print(resultado)
```

```python
resultado1 = re.search("x", texto)
print(resultado1)
```

el metodo .search busca la cadena escrita en la cadena seleccionada y da los indices donde se encuentra.

```python
texto = "Tengo 2 perros y 3 gatos"
resultado = re.findall(r"\d", texto)
print(resultado)
```

El metodo .findall permite encontrar los numeros (\d significa digitos de una cifra) en la cadena.

```python
texto = "Tengo 222 perros y 3 gatos"
resultado = re.findall(r"\d+", texto)
print(resultado)
```

```python
texto = "Python es divertido"
resultado = re.findall(r"\w", texto)
print(resultado)
```

Encuentra letras en una cadena (con \w+ se pueden encontrar palabras y con una lista se pueden encontrar las letras de esta lista. Con \s espacios)

```python
texto = "Me gusta Java"
nuevo = re.sub("Java", "Python", texto)
print(nuevo)
```

el metodo .sub() reemplaza la palabra en la cadena

```python
texto = "rojo,verde,azul"
resultado5 = re.split(",", texto)
print(resultado5)
```

el metodo .split() "corta" la cadena y la convierte en una lista de cadenas.

```python
texto = "Correos: ana@gmail.com y juan@gmail.com"
resultado = re.findall(r"\w+@\w+\.\w+", texto)
print(resultado)
```

Hace una busqueda sencilla de correos electronicos.

```python
telefono = "3022928673"
resultado = re.fullmatch(r"\d{10}", telefono)
print(resultado)
```

Valida una cadena de 10 digitos (por tener {10})

```python
texto = "color colour"
resultado = re.findall(r"colou?r", texto)
print(resultado)
```
