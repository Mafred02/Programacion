```python
dfn = pd.DataFrame({
    'id': [1, 2, 3, 4],
    'nombre': [' Ana', 'LUIS ', 'Carlos', 'ana'],
    'edad': [23, 30, 35, 28],
    'salario': [1000, 2000, 3000, 1500],
    'ciudad': ['Bogotá', 'Cali', 'Bogotá', 'Medellín']
})
dfn
```

```python
dfn['salario_anual'] = dfn['salario'] * 12
dfn['edad_doble'] = dfn['edad'].apply(lambda x:x * 2)
dfn
```

```python
dfn = dfn.rename(columns = {'edad' : 'age'})
dfn
```

```python
dfn['nombre'] = dfn['nombre'].str.lower()
dfn['nombre'] = dfn['nombre'].str.strip()
dfn['nombre'] = dfn['nombre'].str.replace("ana", "JANA")
dfn
```

```python
grupo = dfn.groupby('ciudad')['salario'].mean()
grupo
```

```python
ordenado = dfn.sort_values(by='age')
ordenado
```

```python
dfn1 = pd.DataFrame({
    'id': [1, 2, 3, 4],
    'departamento': ['Ventas', 'IT', 'RRHH', 'Marketing']
})

merge = pd.merge(dfn, dfn1, on="id")
merge
```
