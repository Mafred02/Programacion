```python
#1)
r = np.random.normal(loc=7.2, scale=0.4 , size=52)
d = np.random.gamma(shape=306, scale=0.0229, size=52)
```

```python
#2)
df2 = pd.DataFrame({
    "prueba_A": r,
    "prueba_B": d
})
df2
```

```python
#3)
df2.describe()
```

En la prueba A se obtuvo un promedio aproximado de 7.14, una desviacion de 0.44, una nota minima de 6 y una maxima de 7.98. El 25% saco hasta 6.75, el 50% saco hasta un 7.17 y el 75% saco hasta 7.45.



En la prueba B se obtuvo un promedio aproximado de 7, una desviacion del 0.4 (aproximadamente), un minimo de 6.21 y un maximo de 7.95. El 25% saco hasta un 6.78, el 50% hasta un 7 y el 75% hasta un 7.35.



Con esto podemos concluir que, aunque la diferencia entre los datos obtenidos no es mucha, a los estudiantes en promedio les fue mejor en la Prueba A que en la Prueba B.

```python
#4)
from scipy import stats
stats.shapiro(df2)
```

Como el valor de p es mayor a 0.05, no se rechaza el supuesto de normalidad con una confianza del 95%.

```python
#5)
p_val = stats.ttest_rel(df2["prueba_A"], df2["prueba_B"])
p_val
```

Como el valor de p es mayor a 0.05, no se rechaza que el puntaje medio en ambas encuentas sea igual con una confianza del 95%.

```python
#6)
stats.kendalltau(df2["prueba_A"], df2["prueba_B"])
```

Como el valor de p es menor a 0.05, si se rechaza con una confianza del 95% que los puntajes de ambas encuestas no estan correlacionados.

```python
#7)
stats.spearmanr(df2["prueba_A"], df2["prueba_B"])
```

Como el valor de p es menor a 0.05, si se rechaza con una confianza del 95% que los puntajes de ambas encuestas no estan correlacionados.
