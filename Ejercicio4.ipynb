```python
b = input("Ingrese: ").split(";")
matriz = []
value = True
try:
  for i in range(len(b)):
    a = list(map(float,b[i].split()))
    matriz.append(a)

  for j in range(len(matriz)):
    if len(matriz[j]) != len(matriz[0]):
      value = False
      break
    else:
      continue
  if value == True:
    print(matriz)
  else:
    print("Error")
except ValueError:
  print("Error")
```
